```python
def plot_violin(df, title):
    plt.figure(figsize=(5, 3))
    sns.violinplot(df, palette="viridis", linewidth=0.7)
    plt.ylabel('Energy (kcal/mol)', fontsize=16, fontweight='bold')
    plt.xlabel('Systems', fontsize=16, fontweight='bold')
    plt.title(title, fontweight='bold')
    plt.xticks(rotation=0, fontsize=14)
    plt.yticks(fontsize=12)
    ax = plt.gca()
    ax.spines['left'].set_linewidth(0.7)
    ax.spines['bottom'].set_linewidth(0.7)
    ax.spines['top'].set_linewidth(0.7)
    ax.spines['right'].set_linewidth(0.7)
    plt.tick_params(axis='x', direction='in')
    plt.tick_params(axis='y', direction='in')
    plt.yticks(fontsize=12)
    plt.xticks(rotation=15, fontsize=14)
    #plt.savefig(f'/storage3/gayatrip/figures/mmpbsa/{title.replace(" ", "-").lower()}.svg')
    plt.show()
```

```python
def plot_density(df, title):
    plt.figure(figsize=(7, 5))
    sns.kdeplot(df, palette="viridis", linewidth=0.7, fill=True)
    plt.ylabel('Energy (kcal/mol)', fontsize=16, fontweight='bold')
    plt.xlabel('Systems', fontsize=16, fontweight='bold')
    plt.title(title, fontweight='bold')
    plt.xlabel('Energy (kcal/mol)', fontsize=16, fontweight='bold')
    plt.ylabel('Density', fontsize=16, fontweight='bold')
    plt.xticks(rotation=0, fontsize=14)
    plt.yticks(fontsize=12)
    ax = plt.gca()
    ax.spines['left'].set_linewidth(1)
    ax.spines['bottom'].set_linewidth(1)
    ax.spines['top'].set_linewidth(1)
    ax.spines['right'].set_linewidth(1)
    #plt.legend(fontsize=12)
    # Make the x and y ticks point inward
    plt.tick_params(axis='x', direction='in')
    plt.tick_params(axis='y', direction='in')
    #plt.savefig(f'/storage3/gayatrip/figures/mmpbsa/{title.replace(" ", "-").lower()}.svg')
    plt.show()
```
